### Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
standard_features=scaler.fit_transform(df.drop('quality',axis=1))
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
pc=pca.fit_transform(standard_features)
```

```python
import numpy as np
num_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.9)
```

```python
num_components
```

```python
print("{} of features are required to explain 90% of the variance in the data".format(num_components))
```
